


## 27. Remove Element

**Difficulty:** Easy  
**Topics:** Array, Two Pointers  
**Companies:** (LeetCode)

---

### Problem Statement

Given an integer array `nums` and an integer `val`, remove all occurrences of `val` **in-place**.  
The order of the elements may be changed.

Return the number of elements in `nums` that are **not equal to `val`**.

---

### Requirements

Let `k` be the number of elements in `nums` that are not equal to `val`.

To be accepted, you must:

1. Modify the array `nums` such that the **first `k` elements** contain the elements **not equal to `val`**.
2. The remaining elements of `nums` (after index `k`) are **not important**.
3. Return `k`.

> **Note:**  
> The size of `nums` may change logically, but you must not allocate extra space for another array.

---

If all assertions pass, the solution is accepted.

### Example 1

**Input:**

    nums = [3, 2, 2, 3]
    val = 3

**Output:**

    k = 2
    nums = [2, 2, _, _]

**Explanation:**

*   Elements not equal to `3` are `[2, 2]`
*   First `k = 2` elements are valid
*   Remaining elements can be anything

***

### Example 2

**Input:**

    nums = [0, 1, 2, 2, 3, 0, 4, 2]
    val = 2

**Output:**

    k = 5
    nums = [0, 1, 4, 0, 3, _, _, _]

**Explanation:**

*   Elements not equal to `2` are `[0, 1, 3, 0, 4]`
*   Order does not matter
*   Only first `k = 5` elements are checked

***

### Constraints

*   `0 <= nums.length <= 100`
*   `0 <= nums[i] <= 50`
*   `0 <= val <= 100`



`````python
class Solution:
    def inplace_filter(self , list_of_int  : list[int] , num : int)-> int :
        k=0
        for i in range (len(list_of_int)) :
            if list_of_int[i]!=num :
                list_of_int[k]=list_of_int[i]
                k+=1
        return k , list_of_int
        list_of_int

def main():
    sol= Solution()
    print(sol.inplace_filter([0,1,2,2,4,0,4,2],2))

if __name__ == "__main__":
    main()
`````
