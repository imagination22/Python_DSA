## Problem Statement
Given an array of integers `nums`, return the length of the longest consecutive sequence of elements that can be formed.

A consecutive sequence is a sequence of elements in which each element is exactly 1 greater than the previous element.  
The elements do not have to be consecutive in the original array.

You must write an algorithm that runs in \(O(n)\) time.

---

## Examples

### Example 1
**Input:**  
`nums = [2,20,4,10,3,4,5]`

**Output:**  
`4`

**Explanation:**  
The longest consecutive sequence is `[2, 3, 4, 5]`.

---

### Example 2
**Input:**  
`nums = [0,3,2,5,4,6,1,1]`

**Output:**  
`7`

## Constraints
- \(0 \leq \text{nums.length} \leq 1000\)  
- \(-10^9 \leq \text{nums[i]} \leq 10^9\)


````python
class Merge_sort:
    def merge( self,left:list[int],right )->list[int]:
        i = j = 0
        result = []
        while  (i < len(left)) and j < len(right):
            if left[i] < right[j]:
                result.append(left[i])
                i+=1
                
            else:
                result.append(right[j])
                j+=1
        result.extend(left[i:])
        result.extend(right[j:])
        return result
    def divide( self,arr:list[int])->list[int]: 
        if len(arr) <= 1:
            return arr
        mid = len(arr)//2
        left = self.divide(arr[:mid])
        right = self.divide(arr[mid:])

        return self.merge(left , right)
    
    def merge_sort(self , arr:list[int])->list[int]:
        return self.divide(arr)


class Solution:
    def longest_sequence(self , arr:list[int])->int:
        sorter = Merge_sort()
        sorted_array  = sorter.merge_sort(arr)
        temp_queue = []
        longest_seq = 0
        temp_queue.append(sorted_array[0])
        for i in range(1 , len(sorted_array)):
            if sorted_array[i] == sorted_array[i-1]+1:
                temp_queue.append(sorted_array[i])
                longest_seq +=1
            else:
                longest_seq = max(longest_seq , len(temp_queue))
                
                temp_queue = []
        return longest_seq




def main():
    sol = Solution()
    array = [100, 4, 200, 1, 3, 2,4]
    ans  = sol.longest_sequence(array)
    print(ans)

if __name__ =="__main__":
    main()
````

## ðŸ”¹ Why Itâ€™s O(n), Not O(nÂ²)

At first glance, the nested `while` inside the `for` loop looks quadratic. But hereâ€™s the key:

- **Outer loop:** runs once per element in `arr` â†’ O(n).
- **Inner while loop:** only runs when `num` is the *start of a sequence* (i.e., `num-1 not in set_num`).  
- Each number in the set is visited **at most once** in the inner loop across the entire algorithm.  
  - Example: when starting at `1`, the while loop consumes `1,2,3,4`.  
  - Later, when the outer loop reaches `2`, `3`, or `4`, the condition fails (`num-1 in set_num`), so the while loop is skipped.  

This means the total work done by all while loops combined is **O(n)**, not O(nÂ²).
