# ðŸš€ Solution 1: HashMap (Counting Characters)

## ðŸ’¡ Idea

Count how many times each character appears in both strings.

If counts match â†’ Anagram.

---

## ðŸ”§ Approach

1. Create two hashmaps:

   * `countS`
   * `countT`

2. Count frequency of characters.

3. Compare frequencies.

---

## ðŸ§‘â€ðŸ« Why HashMap?

Because we need:

* Key â†’ Character
* Value â†’ Count of character

Example:

```python

    'a': 3,

    'g': 1,
    'r': 1,
    'm': 1

```

## ðŸ§© Python Implementation

```python
def isAnagram(s, t):
    if len(s) != len(t):
        return False

    countS = {}
    countT = {}

    for i in range(len(s)):
        countS[s[i]] = 1 + countS.get(s[i], 0)
        countT[t[i]] = 1 + countT.get(t[i], 0)

    for c in countS:
        if countS[c] != countT.get(c, 0):
            return False

    return True
```

## ðŸ”Ž Important Concept: `get()` Function

Why not:

```python

```

Because if the key doesn't exist â†’ KeyError âŒ

Instead:

```python

```

Means:

* If key exists â†’ return value
* If not â†’ return 0

Very important interview detail.
