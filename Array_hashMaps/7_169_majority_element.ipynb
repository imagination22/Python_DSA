## Problem Statement  
Given an array `nums` of size `n`, return the majority element.  

The majority element is the element that appears more than âŒŠn / 2âŒ‹ times in the array.  
You may assume that the majority element always exists in the array.

## Follow-up  
Could you solve the problem in **linear time** and in **O(1) space**?

````python

from collections import defaultdict
class Solution:
    def majority_element(self , nums : list[int])->int :
        hashmap = defaultdict(int)
        for num in nums:
            hashmap[num] += 1
        result = max(hashmap,key=hashmap.get)
        if hashmap[result] > len(nums) / 2:
            return result
        return -1

def main():
    sol = Solution()
    print(sol.majority_element([5,5,1,1,1,1,1,1,11,5,5]))
    print(sol.majority_element([2,2,2]))

if __name__ == "__main__":
    main()
````

# ðŸŸ¢ Solution 1 â€” HashMap (Brute but Clean)

## ðŸ’¡ Idea

## ðŸ§± Why HashMap?

Because:

* Values can be arbitrary

* HashMap gives:

  * Insert â†’ O(1)

## ðŸ§® Algorithm Steps

1. Create empty dictionary `count`
2. Maintain:

   * `result` â†’ current best candidate
   * `maxCount` â†’ highest frequency so far
3. Loop through array:

   * Increment count of number
   * If its count > maxCount:

     * Update result
     * Update maxCount
4. Return result

---

## ðŸ’» Code

```python
def majorityElement(nums):
    count = {}
    result = 0
    maxCount = 0

    for n in nums:
        count[n] = 1 + count.get(n, 0)

        if count[n] > maxCount:
            result = n
            maxCount = count[n]

    return result
```

## ðŸ§  Core Insight

## ðŸ§® Algorithm Idea

We maintain:

* `result` â†’ candidate
* `count` â†’ vote counter

Rules:

1. If `count == 0`
   â†’ choose current number as new candidate

2. If current number == candidate

3. Else

## ðŸ§  Why It Works

Think of it as **pair cancellation**.

Every time we see a different number:

* We cancel one vote of the current candidate.

Since majority element appears more than half:

* It will survive all cancellations.

## ðŸ’» Code

```python
def majorityElement(nums):
    result = 0
    count = 0

    for n in nums:
        if count == 0:
            result = n

        if n == result:
            count += 1
        else:
            count -= 1

    return result
```
