# Top K Frequent Elements

**Difficulty:** Medium  
**Topics:** Hashing, Heap/Priority Queue, Sorting  
**Company Tags:** Common interview problem (Google, Amazon, Microsoft, etc.)  
**Hints:** Use a frequency map and a heap or bucket sort.

---

## Problem Statement
Given an integer array `nums` and an integer `k`, return the `k` most frequent elements within the array.

The test cases are generated such that the answer is always unique.

You may return the output in any order.

---

## Examples

### Example 1
**Input:**  
`nums = [1,2,2,3,3,3], k = 2`  

**Output:**  
`[2,3]`

---

### Example 2
**Input:**  
`nums = [7,7], k = 1`  

**Output:**  
`[7]`

---

## Constraints
- \(1 \leq \text{nums.length} \leq 10^4\)  
- \(-1000 \leq \text{nums[i]} \leq 1000\)  
- \(1 \leq k \leq\) number of distinct elements in `nums`

---

## Approach Ideas
1. **Hash Map + Heap (Priority Queue)**  
   - Count frequencies using a dictionary.  
   - Use a max-heap or min-heap to extract top `k` frequent elements.  
   - Time Complexity: \(O(N \log k)\).  
   - Space Complexity: \(O(N)\).

2. **Hash Map + Bucket Sort**  
   - Count frequencies.  
   - Place elements into buckets indexed by frequency.  
   - Collect top `k` elements from highest frequency buckets.  
   - Time Complexity: \(O(N)\).  
   - Space Complexity: \(O(N)\).

---

## Key Insight
The problem is essentially about **frequency counting** and then **selecting the top k elements**.  
Efficient solutions avoid sorting the entire array and instead focus only on the frequency distribution.


````python
from collections import Counter
class Solution:
    def most_freq(arr : list[int] , num :int) -> list[int]:
        hash_map = Counter(arr)
        result = []
        for i in range (num):
            result.append(hash_map.most_common(num)[i][0])
        return result
    
arr = [1,1,1,2,2,3]
num = 2
print(Solution.most_freq(arr,num))
````

## ðŸ”¹ Step 3: Understanding `hash_map.most_common(num)[i][0]`
Letâ€™s break it down:
- `hash_map.most_common(num)` â†’ returns a list of tuples.  
  Example: `[(1, 3), (2, 2)]`.  
- `[i]` â†’ picks the i-th tuple.  
  Example: for `i=0`, itâ€™s `(1, 3)`.  
- `[0]` â†’ picks the first element of the tuple (the actual number, not the frequency).  
  Example: `(1, 3)[0] = 1`.

So:
```python

```
means:  
**â€œGive me the element (not the count) of the i-th most frequent item.â€**

````python
class Solution:
    def topKFrequent(self, nums: list[int], k: int) -> list[int]:
        count = {}
        freq = [[] for i in range(len(nums) +1)]
        print(freq)

        for n in nums:
            count[n] =1+ count.get(n, 0)
        for n, c in count.items():
            freq[c].append(n)

        res = []
        for i in range(len(freq) - 1, 0, -1):
            for n in freq[i]:
                res.append(n)
            if len(res) == k:
                return res

sol = Solution()
print(sol.topKFrequent([1,1,1,2,2,3],2))
````
