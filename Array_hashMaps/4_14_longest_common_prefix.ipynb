# Longest Common Prefix
**Difficulty:** Easy  
**Topics:** Strings  

---

### Problem
You are given an array of strings `strs`.  
Return the longest common prefix of all the strings.  

If there is no longest common prefix, return an empty string `""`.

---

### Example 1
**Input:**  
`strs = ["bat","bag","bank","band"]`  

**Output:**  
`"ba"`  

---

### Example 2
**Input:**

`"da"`  

---

### Example 3
**Input:**

**Output:**  
`""`  

---

### Constraints
- `1 <= strs.length <= 200`  
- `0 <= strs[i].length <= 200`  
- `strs[i]` is made up of lowercase English letters if it is non-empty.

# ðŸ’» Python Implementation

```python
def longestCommonPrefix(strs):
    if not strs:
        return ""

    result = ""

    for i in range(len(strs[0])):  # iterate through first string
        for s in strs:
            # Check out of bounds or mismatch
            if i == len(s) or s[i] != strs[0][i]:
                return result
        
        result += strs[0][i]

    return result
```
