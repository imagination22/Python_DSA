### Problem
Given an integer array `nums` of length `n`, you want to create an array `ans` of length `2n` where:

- `ans[i] == nums[i]`  
- `ans[i + n] == nums[i]` for `0 <= i < n`  

Specifically, `ans` is the concatenation of two `nums` arrays.  

Return the array `ans`.

---

### Example 1
**Input:**  
`nums = [1,2,1]`  

**Output:**  
`[1,2,1,1,2,1]`  

**Explanation:**  
- `ans = [nums[0], nums[1], nums[2], nums[0], nums[1], nums[2]]`  
- `ans = [1,2,1,1,2,1]`  

---

### Example 2
**Input:**  
`nums = [1,3,2,1]`  

**Output:**  
`[1,3,2,1,1,3,2,1]`  

**Explanation:**  
- `ans = [nums[0], nums[1], nums[2], nums[3], nums[0], nums[1], nums[2], nums[3]]`  
- `ans = [1,3,2,1,1,3,2,1]`  

---

### Constraints
- `n == nums.length`  
- `1 <= n <= 1000`  
- `1 <= nums[i] <= 1000`

Our goal is to solve a simple but important problem:

> Given an integer array `nums` of length `n`, return its concatenation with itself.

---

## ðŸ§  Understanding the Problem

Suppose we are given:

```python
nums = [1, 2, 1]
```

Length `n = 3`.

We want to return:

```python
[1, 2, 1, 1, 2, 1]
```

What did we do?

We took the original array
and appended it to itself.

So conceptually:

```
result = nums + nums
```

Thatâ€™s it.

Two copies combined into one single array.



````python
class Solution:
    def getConcatenation_sol_1(self, nums: list[int]) -> list[int]:
        ans = nums.copy()
        ans.extend(nums)
        return ans
    def getConcatenation_sol_2(self, nums: list[int]) -> list[int]:
        ans = []
        for _ in range (2):
            for i in nums :
                ans.append(i)
        return ans
    def getConcatenation_sol_3(self, nums: list[int]) -> list[int]:
        return nums + nums
    def getConcatenation_sol_4(self, nums: list[int]) -> list[int]:
        return nums *2
    
    
 

def main():
    sol = Solution()
    result_1 = sol.getConcatenation_sol_1([1,2,3,4])
    result_2 = sol.getConcatenation_sol_2([1,2,3,4])
    result_3 = sol.getConcatenation_sol_3([1,2,3,4])
    result_4 = sol.getConcatenation_sol_3([1,2,3,4])
    print(result_1)
    print(result_2)
    print(result_3)
    print(result_4)

if __name__ == "__main__":
    main()
````

# ðŸ§© Approach 2 â€” Create a New Result Array (Cleaner & Extensible)

This is the better design.

### Step 1: Create empty result array

```python
ans = []
```

### Step 2: Add elements from nums

Loop through `nums` and append each element to `ans`.

But remember:

We need **two copies**, not one.

So we repeat the process twice.

This approach is:

* Cleaner

# ðŸ§® Time Complexity Analysis

# ðŸ’» Final Code (Clean Version)

```python
class Solution:
    def getConcatenation(self, nums):
        ans = []

        for _ in range(2):              # run twice
            for num in nums:
                ans.append(num)

        return ans
```
