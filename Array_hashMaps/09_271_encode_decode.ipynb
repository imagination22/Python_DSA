## Problem Statement
Design an algorithm to encode a list of strings into a single string.  
The encoded string is then sent over the network and decoded back to the original list of strings.

- **Machine 1 (Sender)** has the function:
  ```cpp
  string encode(vector<string> strs) {
      // ... your code
      return encoded_string;
  }

- **Machine 2 (Reciever)** has the function:
vector<string> decode(string s) {
    // ... your code
    return strs;
}
So Machine 1 does:

cpp
string encoded_string = encode(strs);
And Machine 2 does:

cpp
vector<string> strs2 = decode(encoded_string);
strs2 in Machine 2 should be the same as strs in Machine 1.

Implement the encode and decode methods.

Examples
Example 1
Input:  
dummy_input = ["Hello","World"]

Output:  
["Hello","World"]

Explanation:

Machine 1 encodes the list into a single string.

Machine 2 decodes it back into the original list.

The result matches the input.

Example 2
Input:  
dummy_input = [""]

Output:  
[""]

Constraints
0
≤
strs.length
<
100

0
≤
strs[i].length
<
200

strs[i] contains any possible characters out of 256 valid ASCII characters.

Follow-up
Could you write a generalized algorithm to work on any possible set of characters?
Hint: Consider encoding lengths of strings explicitly to avoid delimiter conflicts.


````python
class Solution:
    def encoder(self,input_list : list[str])->str:
        str_result = ''
        for i in input_list:
            str_result += str(len(i))+ '#' + i
        return str_result
        
    def decoder(self,encoded_str : str)->list[str]:
 
        decode_list = []
        i=0
        while i < len(encoded_str):
            j=i
            while encoded_str[j] != '#':
                j+=1
            len_s =int(encoded_str[i:j])
            decode_list.append(encoded_str[j+1:j+1+len_s])
            i = j+1+len_s
        return decode_list


if __name__ =="__main__":
    sol = Solution()
    q_1_list = ['hello','my','name','is','khan','and' ,'i','am','not','a','terrorist']
    print(q_1_list)
    endcoded_str = sol.encoder(q_1_list)
    print(endcoded_str)
    decoded_list = sol.decoder(endcoded_str)
    print(decoded_list)
````
