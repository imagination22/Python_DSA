```python
class Node:
    def __init__(self , val):
        self.data = val
        self.prev = None
        self.next = None

```

```python
class Stack:
    def __init__(self ):
        self.head = None
        self.count = 0
    def push(self , val):
        self.data = Node(val)
        if self.head is not  None :
           self.data.prev = self.head
        self.head = self.data
        self.count +=1

    def pop(self ):
        if self.head is None:
            return None
        else :
            self.head = self.data.prev
            self.count -=1
            return self.data.data
    def peek(self):
        return  self.head.data
    
    def traverse_and_print(self):
        pointer = self.head
        while pointer is not None:
            print(pointer.data,end ="-->")
            pointer = pointer.prev
            if pointer is  None:
                print("None")



```

```python
if __name__ =="__main__":
    stack = Stack()
    stack.push('a')
    stack.push('b')
    stack.push('c')
    stack.push('d')
    stack.push('e')
    stack.push('f')
    stack.traverse_and_print()
    value =  stack.peek()
    print(value)
    value = stack.pop()
    print(value)
    stack.traverse_and_print()
```
