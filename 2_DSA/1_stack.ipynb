## ðŸ”¹ What is a Stack?
A **stack** is a linear data structure that follows the **LIFO (Last In, First Out)** principle:
- The last element inserted is the first one removed.
- Think of a stack of plates: you add on top, and you remove from the top.

---

## ðŸ”¹ Basic Operations
| Operation | Description | Time Complexity |
|-----------|-------------|-----------------|
| **Push(x)** | Insert element `x` at the top | \(O(1)\) |
| **Pop()** | Remove the top element | \(O(1)\) |
| **Peek()/Top()** | View the top element without removing | \(O(1)\) |
| **isEmpty()** | Check if stack is empty | \(O(1)\) |
| **isFull()** (for fixed-size stack) | Check if stack is full | \(O(1)\) |

Stacks can be implemented using:
- **Arrays** (fixed size, simple).  
- **Linked Lists** (dynamic size, more flexible).

## ðŸ”¹ Stack Example
```python
class Stack:
    def __init__(self):
        self.stack = []   # each object gets its own list

    def push(self, element):
        self.stack.append(element)

    def pop(self):
        if self.isEmpty():
            return "Stack is empty"
        return self.stack.pop()

    def peek(self):
        if self.isEmpty():
            return "Stack is empty"
        return self.stack[-1]

    def isEmpty(self):
        return len(self.stack) == 0

    def size(self):
        return len(self.stack)
```

### What happens when you do:
```python
myStack = Stack()
```
- Python calls `__init__`.  
- Inside `__init__`, it creates `self.stack = []`.  
- So now `myStack` has its own empty list.  

When you call:
```python
myStack.push("A")
```
- It runs the `push` method on this object.  
- `self.stack.append("A")` adds `"A"` to the list inside `myStack`.  

Each object you create (`Stack()`) has its own separate list because `self` points to that specific object.

## ðŸ”¹ Passing Values During Initialization
You can also design a class to accept values directly when you create it:

```python
class Stack:
    def __init__(self, initial=None):
        if initial is None:
            self.stack = []
        else:
            self.stack = initial

# Example
s1 = Stack(["A", "B"])   # starts with A, B already inside
s2 = Stack()             # starts empty
```

Here, the argument `initial` is passed into the constructor, and `self.stack` stores it. Thatâ€™s how values can be passed directly during object creation.

## ðŸ”¹ â€œnextâ€ Variable Storing Next Object
This comes into play in **linked structures** (like linked lists or linked stacks). Instead of using arrays, you chain nodes together:

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None   # will point to another Node object

# Example
n1 = Node("A")
n2 = Node("B")
n1.next = n2   # n1 points to n2
```

Here:
- `n1` stores `"A"`.  
- `n2` stores `"B"`.  
- `n1.next` points to the **object `n2`**.  

So `next` is just a variable that holds a reference to another object in memory. Thatâ€™s how linked stacks or linked lists are built â€” each node points to the next one.

---

## ðŸ”¹ Connecting Back to Stack
- In your current stack, you used a **list-based implementation** (Pythonâ€™s built-in list).  
- In a **linked-list stack**, each node has a `data` and a `next`. The stack keeps track of the â€œtopâ€ node.  
- When you `push`, you create a new node and link it.  
- When you `pop`, you move the â€œtopâ€ pointer to the next node.

---

- **Initialization arguments** â†’ passed into `__init__` and stored in `self`.  
- **next variable** â†’ just a pointer/reference to another object, used in linked structures.  
- OOP makes this neat because each object manages its own data and relationships.

````python
class Stack:
  def __init__(self):
    self.stack = []

  def push(self, element):
    self.stack.append(element)

  def pop(self):
    if self.isEmpty():
      return "Stack is empty"
    return self.stack.pop()

  def peek(self):
    if self.isEmpty():
      return "Stack is empty"
    return self.stack[-1]

  def isEmpty(self):
    return len(self.stack) == 0

  def size(self):
    return len(self.stack)

# Create a stack
myStack = Stack()

myStack.push('A')
myStack.push('B')
myStack.push('C')

print("Stack: ", myStack.stack)
print("Pop: ", myStack.pop())
print("Stack after Pop: ", myStack.stack)
print("Peek: ", myStack.peek())
print("isEmpty: ", myStack.isEmpty())
print("Size: ", myStack.size())
````

````python
class Node:
  def __init__(self, value):
    self.value = value
    self.next = None

class Stack:
  def __init__(self):
    self.head = None
    self.size = 0

  def push(self, value):
    new_node = Node(value)
    if self.head:
      new_node.next = self.head
    self.head = new_node
    self.size += 1

  def pop(self):
    if self.isEmpty():
      return "Stack is empty"
    popped_node = self.head
    self.head = self.head.next
    self.size -= 1
    return popped_node.value

  def peek(self):
    if self.isEmpty():
      return "Stack is empty"
    return self.head.value

  def isEmpty(self):
    return self.size == 0

  def stackSize(self):
    return self.size

  def traverseAndPrint(self):
    currentNode = self.head
    while currentNode:
      print(currentNode.value, end=" -> ")
      currentNode = currentNode.next
    print()

myStack = Stack()
myStack.push('A')
myStack.push('B')
myStack.push('C')

print("LinkedList: ", end="")
myStack.traverseAndPrint()
print("Peek: ", myStack.peek())
print("Pop: ", myStack.pop())
print("LinkedList after Pop: ", end="")
myStack.traverseAndPrint()
print("isEmpty: ", myStack.isEmpty())
print("Size: ", myStack.stackSize())
````
