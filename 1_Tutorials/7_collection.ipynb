# ğŸ“¦ 1. `namedtuple`

### âœ… What it is

A lightweight object type that behaves like a tuple **but with named fields**.

### ğŸ“Œ Why use it?

* More readable than normal tuple
* Immutable
* Memory efficient
* Good alternative to small classes

---

### ğŸ”¹ Syntax

```python
from collections import namedtuple

Point = namedtuple("Point", ["x", "y"])
p = Point(10, 20)

print(p.x)   # 10
print(p[0])  # 10
```

---

### ğŸ”¹ Use Case

Instead of:

```python
user = ("John", 25)
print(user[0])
```

Use:

```python
User = namedtuple("User", ["name", "age"])
u = User("John", 25)
print(u.name)
```

âœ” Cleaner
âœ” Self-documenting

---

### âŒ When NOT to use

* If you need mutability
* If you need complex methods (use class instead)

# ğŸ“¦ 2. `deque` (Double Ended Queue)

### âœ… What it is

A list-like container optimized for **fast append/pop from both ends**.

---

### ğŸ”¹ Why use it?

| Operation    | list | deque |
| ------------ | ---- | ----- |
| append right | fast | fast  |
| append left  | slow | fast  |
| pop right    | fast | fast  |
| pop left     | slow | fast  |

---

### ğŸ”¹ Syntax

```python
from collections import deque

d = deque([1,2,3])

d.append(4)       # right
d.appendleft(0)   # left
d.pop()
d.popleft()
```

---

### ğŸ”¹ Special Features

```python
d.rotate(1)       # rotate right
d.rotate(-1)      # rotate left
```

```python
d = deque([1,2,3], maxlen=3)
d.append(4)
print(d)   # oldest removed automatically
```

---

### ğŸ”¹ Use Cases

âœ” Implement stack
âœ” Implement queue
âœ” Sliding window problems
âœ” LRU cache logic

---

### âŒ When NOT to use

* When random indexing is primary need (list is better)

# ğŸ“¦ 3. `Counter`

### âœ… What it is

### ğŸ”¹ Syntax

### ğŸ”¹ Useful Methods

---

### ğŸ”¹ Use Cases

âœ” Comparing inventories

---

### âŒ When NOT to use

* If counting is not required

# ğŸ“¦ 4. `defaultdict`

### âœ… What it is

Dictionary that provides a default value for missing keys.

---

### ğŸ”¹ Normal dict problem

```python
d = {}
d["a"].append(1)  # KeyError
```

---

### ğŸ”¹ With defaultdict

```python
from collections import defaultdict

d = defaultdict(list)
d["a"].append(1)
print(d)
```

---

### ğŸ”¹ Common Patterns

```python
defaultdict(list)
defaultdict(int)
defaultdict(set)
```

---

### ğŸ”¹ Use Cases

âœ” Grouping data
âœ” Counting
âœ” Graph adjacency list

---

### âŒ When NOT to use

* When missing keys should raise error

# ğŸ“¦ 5. `OrderedDict`

### âœ… What it is

### ğŸ”¹ Why still useful?

It provides extra methods:

```python
from collections import OrderedDict

d = OrderedDict()
d["a"] = 1
d.move_to_end("a")
d.popitem(last=False)
```

### ğŸ”¹ Use Case

### âŒ When NOT needed

# ğŸ“¦ 6. `ChainMap`

### âœ… What it is

Combines multiple dictionaries into one logical view.

---

### ğŸ”¹ Example

```python
from collections import ChainMap

d1 = {"a":1}
d2 = {"b":2}

c = ChainMap(d1, d2)
print(c["a"])
print(c["b"])
```

---

### ğŸ”¹ Use Cases

âœ” Managing layered configs
âœ” Variable scope simulation
âœ” Overriding defaults

---

### âŒ When NOT to use

* If you need merged dictionary (use `{**d1, **d2}`)

# ğŸ“¦ 7. `UserDict`, `UserList`, `UserString`

### âœ… What they are

Wrapper classes for easier subclassing of built-ins.

---

### ğŸ”¹ Why use them?

Instead of subclassing dict directly:

```python
from collections import UserDict

class MyDict(UserDict):
    def __setitem__(self, key, value):
        print("Setting:", key)
        super().__setitem__(key, value)
```

---

### ğŸ”¹ Use Cases

âœ” Custom validation
âœ” Logging
âœ” Behavior modification

---

### âŒ When NOT needed
