# 6ï¸âƒ£ for-else (HIGH CONFUSION)

Structure:

```python
for item in iterable:
    if condition:
        break
else:
    print("No break happened")
```

Important:

* else runs ONLY if loop finishes normally
* else does NOT run if break occurs

âš  else is attached to loop, not if.

# 7ï¸âƒ£ match (Python 3.10+)

Replacement for long if-elif chains.

```python
match value:
    case 1:
        ...
    case 2:
        ...
    case _:
        ...
```

* `_` is default
* Can match patterns

Example:

```python
match x:
    case 1 | 2:
        print("One or two")
    case _:
        print("Other")
```
